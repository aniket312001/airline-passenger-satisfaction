# Missing values

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train.fillna({'Arrival Delay in Minutes':df_train['Arrival Delay in Minutes'].median()},inplace=True)
df_test.fillna({'Arrival Delay in Minutes':df_test['Arrival Delay in Minutes'].median()},inplace=True)
```

```python
df_train.info()
```

# Label Encoding

```python
categories = [feat for feat in df_train.columns  if df_train[feat].dtype =='O']
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

for feat in categories: 
    df_train[feat] = label.fit_transform(df_train[feat])
    
```

```python
for feat in categories:
    df_test[feat] = label.fit_transform(df_test[feat])
```

```python
df_train.head()
```

```python
df_test.head()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scale_train = scaler.fit_transform(df_train.drop(['satisfaction','Unnamed: 0','id'],axis=1))
scale_test = scaler.fit_transform(df_test.drop(['satisfaction','Unnamed: 0','id'],axis=1))
```

```python
scale_train = pd.DataFrame(scale_train,columns=df_train.columns[2:-1])
scale_test = pd.DataFrame(scale_test,columns=df_test.columns[2:-1])
```

```python
scale_train.head()
```

```python
scale_test.head()
```

```python
y_train = df_train['satisfaction']
y_test = df_test['satisfaction']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

lg = LogisticRegression()
lg.fit(scale_train,y_train)

lg_predict = lg.predict(scale_test)


print(classification_report(y_test,lg_predict))
print()
print(confusion_matrix(y_test,lg_predict))
print()
print('Accuracy -> ',accuracy_score(y_test,lg_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(scale_train,y_train)

rf_predict = rf.predict(scale_test)


print(classification_report(y_test,rf_predict))
print()
print(confusion_matrix(y_test,rf_predict))
print()
print('Accuracy -> ',accuracy_score(y_test,rf_predict))
```
