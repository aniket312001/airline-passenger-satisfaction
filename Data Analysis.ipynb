# Numeric Variable

```python
Numeric = [feat for feat in df_train.columns if df_train[feat].dtype != 'O' and feat not in 'id' + 'Unnamed: 0' ]

df_train[Numeric].head()
```

 There are two type numeric variable
    they are discrete and continous

```python
discrite = [feat for feat in Numeric if len(df_train[feat].unique())<20 ]

df_train[discrite].head()
```

```python
for feat in discrite:
    sns.countplot(df_train[feat],hue=df_train['satisfaction'])
    plt.title(feat)
    plt.show()
```

```python
continous = [feat for feat in Numeric if feat not in discrite]
df_train[continous].head()
```

```python
for feat in continous:
    df_train[feat].plot.hist(bins=30)
    plt.title(feat)
    plt.show()
```

# Categorical Variable

```python
categories = [feat for feat in df_train.columns  if df_train[feat].dtype =='O']
df_train[categories].head()
```

```python
for feat in categories:
    print(f'No. of features in {feat} is {len(df_train[feat].unique())}')
```

```python
for feat in categories:
    sns.countplot(df_train[feat],hue=df_train['satisfaction'])
    plt.title(feat)
    plt.show()
```
